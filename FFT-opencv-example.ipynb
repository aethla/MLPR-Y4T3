```python
# First import the libraries cv2 and matplotlib
import cv2
import numpy as np
import matplotlib.pyplot as plt
path = "/content/drive/MyDrive/MLPR - Y4T3/messi.jpg"
# Read an image file using cv2
img = cv2.imread(path, 0)
type(img)
```

```python
plt.figure(figsize=(10,5))
plt.imshow(img)
plt.title('origin')
plt.show()
```

```python
f = np.fft.fft2(img)
fs = np.fft.fftshift(f)
lfs = 15 * np.log(1 + np.abs(fs))
lfs = np.uint8(lfs)
```

```python
plt.figure(figsize=(10,5))
plt.imshow(lfs)
plt.title('dft')
plt.show()
```

```python
#reversing fourier transform process
ishift_img = np.fft.ifftshift(fs)
ifft_img = np.fft.ifft2(ishift_img)
abs_img = np.abs(ifft_img)
abs_img =np.uint8(abs_img)
plt.figure(figsize=(10,5))
plt.imshow(abs_img )
plt.title('idft')
plt.show()
```
